```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\marco\Desktop\ENTREVISTA DOCENTE.xlsx")
fila ="¿Qué herramientas tecnológicas ha requerido para prevenir los posibles plagios en los exámenes online (Seleccione las que crea conveniente)"
df.rename(columns={f"{fila}": "pregunta multiple"}, inplace=True)

df = df[['pregunta multiple']].dropna()

conteo_problemas = df.value_counts().reset_index()
conteo_problemas.columns = ["Problema", "Frecuencia"]

plt.figure(figsize=(22, 10))
plt.grid(True, linestyle="--", color="black", alpha=0.5) 

barplot = sns.barplot(data=conteo_problemas, y="Problema", x="Frecuencia", palette="viridis")

for index, value in enumerate(conteo_problemas["Frecuencia"]):
    barplot.text(value + 0.1, index, str(value), color="black", ha="center", va="center", fontsize=12, weight="bold")

plt.title("""¿Qué herramientas tecnológicas ha requerido para prevenir los posibles plagios 
          en los exámenes online (Seleccione las que crea conveniente)""")
plt.xlabel("Frecuencia", fontsize=14, color="gray")
plt.ylabel("Tipo de Problema", fontsize=14, color="gray")
plt.xticks(fontsize=12, color="black")
plt.yticks(fontsize=12, color="black")

plt.tight_layout()

plt.show()
```

```python
df_separado = df["pregunta multiple"].str.split(",", expand=True)

problemas_separados = df_separado.stack().reset_index(drop=True)

conteo_problemas = problemas_separados.value_counts().reset_index()
conteo_problemas.columns = ["Problema", "Frecuencia"]

plt.figure(figsize=(14, 10))
plt.grid(True, linestyle="--", color="black", alpha=0.5) 

barplot = sns.barplot(data=conteo_problemas, y="Problema", x="Frecuencia", palette="viridis")

for index, value in enumerate(conteo_problemas["Frecuencia"]):
    barplot.text(value + 0.5, index, str(value), color="black", ha="center", va="center", fontsize=12, weight="bold")

plt.xlabel("Frecuencia", fontsize=14, color="gray")
plt.ylabel("Tipo de Problema", fontsize=14, color="gray")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.show()
```
